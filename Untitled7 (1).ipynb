```python
import pandas as pd
import sqlite3
```

```python
from sqlalchemy import create_engine
engine = create_engine('sqlite://' ,echo = False)
```

```python
covid = pd.read_excel('C:/Users/Sam/Downloads/covid-data.xlsx')
vaccine= pd.read_excel('C:/Users/Sam/Downloads/vaccinedata.xlsx')
covid.head(50)

```

```python
print(covid.columns.to_list())
```

```python
covid.to_sql('coviddata', con = engine)
vaccine.to_sql('vaccine1data', con = engine)
```

```python
sql =  '''

select * from coviddata
where continent is null

'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(10)
```

```python
sql = '''

Select Location, date, total_cases,total_deaths, (total_deaths/total_cases)*100 as DeathPercentage
From coviddata
Where location like '%states%'
and continent is not null 
order by 1,2

'''
df_sql = pd.read_sql_query(sql, con = engine)
df_sql.head(50)
```
